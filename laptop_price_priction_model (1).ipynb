```python
import gradio as gr
import pandas as pd
import pickle

df = pd.read_csv('laptop.csv')
pipe = pickle.load(open('pipe.pkl','rb'))
```

```python
def predict_price(Company,TypeName,Ram,Weight,Touchscreen,IPS,ppi,Cpu_brand,HDD,SSD,Gpu_brand,os):
  input_df = pd.DataFrame([[Company,TypeName,Ram,Weight,Touchscreen,IPS,ppi,Cpu_brand,HDD,SSD,Gpu_brand,os]],columns=['Company','TypeName','Ram','Weight','Touchscreen','IPS','ppi','Cpu brand','HDD','SSD','Gpu brand','os'])
  
  # Preprocess the input
  Company = input_df['Company'].map({'Apple':0,'Razer':1,'Mediacom':2,'Samsung':3,'Toshiba':4,'MSI':5,'Vero':6,'Dell':7,'Huawei':8,'Chuwi':9,'Fujitsu':10,'LG':11,'Google':12,'HP':13,'Asus':14,'Acer':15,'Lenovo':16,'Nokia':17,'Microsoft':18})
  input_df['Company'] = Company

  TypeName = input_df['TypeName'].map({'Ultrabook':0,'Notebook':1,'Netbook':2,'Gaming':3,'2 in 1 Convertible':4,'Workstation':5})
  input_df['TypeName'] = TypeName

  Cpu_brand = input_df['Cpu brand'].map({'Intel Core i7':0,'Intel Core i5':1,'Intel Core i3':2,'Other Intel Processor':3,'AMD Processor':4})
  input_df['Cpu brand'] = Cpu_brand

  Gpu_brand = input_df['Gpu brand'].map({'Intel':0,'Nvidia':1,'AMD':2,'ARM':3})
  input_df['Gpu brand'] = Gpu_brand

  os = input_df['os'].map({'Mac':0,'Others/No OS/Linux':1,'Windows':2})
  input_df['os'] = os
  
  prediction = np.exp(pipe.predict(input_df))
  return prediction[0]
```

```python
# Define the input components for the interface
new_var = [
    gr.Dropdown(choices=list(df['Company'].unique()), label="Company"),
    gr.Dropdown(choices=list(df['TypeName'].unique()), label="TypeName"),
    gr.Slider(4, 64, step=4, label="Ram"),
    gr.Slider(0.5, 5.0, step=0.1, label="Weight"),
    gr.Radio(choices=[0, 1], label="Touchscreen"),
    gr.Radio(choices=[0, 1], label="IPS"),
    gr.Slider(50, 500, step=1, label="ppi"),
    gr.Slider(0, 2000, step=1, label="HDD"),
    gr.Slider(0, 2000, step=1, label="SSD"),    
    gr.Dropdown(choices=list(df['OpSys'].unique()), label="os") # Changed 'os' to 'OpSys'
]
inputs = new_var

# Define the output component for the interface
outputs = gr.Number(label="Price")

# Create the Gradio interface
iface = gr.Interface(fn=predict_price, inputs=inputs, outputs=outputs)

# Launch the interface
iface.launch()
share=True
```
